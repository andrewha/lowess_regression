```python
import sys
sys.path.append('../lib')
import numpy as np
import matplotlib.pyplot as plt
from lowess_regression import LowessModel
```

```python
#np.random.seed(56) # Comment this line to get new result on every run
n = 200
# Original heart shape
t = np.sort(np.random.uniform(-np.pi/10 * np.pi, np.pi/10 * np.pi, size=n))
x = np.log(np.abs(t)) * np.sin(t) * np.cos(t)
y = np.sqrt(np.abs(t)) * np.cos(t) 
```

```python
# Distort the signal by introducing some additive noise
#noise = 'white'
noise = 'gaussian'
y += np.random.uniform(0, 0.25, size=len(y)) if noise == 'white' else np.random.normal(0, 0.05, size=len(y))
```

```python
# Set model's params
kernel_param = 0.25
n_std = 1.0
```

```python
# Create and apply our model to distorted signal
lf = LowessModel(kernel_param, n_std)
y_smoothed = lf.smooth(x, y)
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(x, y, marker='+', color='C0')
plt.scatter(x, y_smoothed[0], color='brown', alpha=0.5)
plt.axis('off');
```
